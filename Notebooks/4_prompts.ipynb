```python
from dotenv import load_dotenv
load_dotenv()

import os

if os.environ['GROQ_API_KEY']:
    print("API Key is Set!!!")
else:
    raise ValueError("API Key not loaded!!!")
```

```python
from langchain_groq import ChatGroq
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")

llm = ChatGroq(model="openai/gpt-oss-120b")
llm
```

```python
from langchain_core.prompts import ChatPromptTemplate

my_prompt = ChatPromptTemplate.from_messages(
    [
        ("system","You are a helpful assistant that helps to create and curate posts for linkedin"),
        ("human", "I want to create a post on {topic}")
    ]
)
```

```python
prompt = my_prompt.invoke({"topic":"What is the importance of AI in present world?"})
```

```python
llm.invoke(prompt)
```

### **Chain with prompts**

```python
chain = my_prompt | llm

chain.invoke({"topic":"What is the importance of AI in present world?"})
```
